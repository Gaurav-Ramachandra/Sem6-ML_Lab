```python
# Sales Prediction

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Load the dataset
# url = "https://raw.githubusercontent.com/your-repo/sales_data.csv"  # Replace with actual URL if hosted online
# df = pd.read_csv(url)
df = pd.read_csv("/content/drive/MyDrive/Sem-6/ML/Lab1/sales_data.csv")

# Split data into independent (X) and dependent (Y) variables
X = df[['Day']].values  # Independent variable (Days)
Y = df[['Sales ($1000)']].values  # Dependent variable (Sales)

# Train the Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict sales
Y_pred = model.predict(X)

# Model parameters
print(f"Intercept: {model.intercept_[0]:.2f}")
print(f"Coefficient: {model.coef_[0][0]:.2f}")

# Calculate errors
mae = mean_absolute_error(Y, Y_pred)
mse = mean_squared_error(Y, Y_pred)
rmse = np.sqrt(mse)

print(f"Mean Absolute Error: {mae:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")

# Plot the actual vs predicted values
plt.figure(figsize=(8, 5))
plt.scatter(X, Y, color='blue', label="Actual Sales")
plt.plot(X, Y_pred, color='red', linewidth=2, label="Predicted Sales")
plt.xlabel("Day")
plt.ylabel("Sales ($1000)")
plt.title("Linear Regression: Sales Prediction")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Salary Prediction

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Sem-6/ML/Lab1/Salary_dataset.csv")  # Ensure file is uploaded in Colab

# Display first few rows to verify column names
print(df.head())

# Drop the unnecessary "Unnamed: 0" column if present
df = df.drop(columns=["Unnamed: 0"], errors='ignore')

# Use correct column names
X = df[['YearsExperience']].values  # Independent variable
Y = df[['Salary']].values  # Dependent variable

# Train the Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict salaries
Y_pred = model.predict(X)

# Model parameters
print(f"Intercept: {model.intercept_[0]:.2f}")
print(f"Coefficient: {model.coef_[0][0]:.2f}")

# Calculate errors
mae = mean_absolute_error(Y, Y_pred)
mse = mean_squared_error(Y, Y_pred)
rmse = np.sqrt(mse)

print(f"Mean Absolute Error: {mae:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")

# Plot actual vs predicted salaries
plt.figure(figsize=(8, 5))
plt.scatter(X, Y, color='blue', label="Actual Salary")
plt.plot(X, Y_pred, color='red', linewidth=2, label="Predicted Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression: Salary Prediction")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# deekshith nologin code for Salary Prediction

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("/content/drive/MyDrive/Sem-6/ML/Lab1/Salary_dataset.csv")

# Limit data to 40-50 points (adjustable)
df = df.iloc[:50]  # Select first 50 rows (change as needed)

# Assuming first column as X and second as y
X = df.iloc[:, 0].values
y = df.iloc[:, 1].values

# Compute linear regression parameters using the least squares formula
X_mean, y_mean = np.mean(X), np.mean(y)

# Slope (m) and Intercept (b)
m = np.sum((X - X_mean) * (y - y_mean)) / np.sum((X - X_mean) ** 2)
b = y_mean - m * X_mean

# Predicted values
y_pred = m * X + b

# Calculate Mean Squared Error (MSE)
mse = np.mean((y - y_pred) ** 2)
print(f"Slope (m): {m}, Intercept (b): {b}, MSE: {mse}")

# Plot data points and regression line
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.title("Linear Regression (Limited to 50 Points)")
plt.show()
```

```python
# General Regression- works for both simple and multiple

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Load the dataset (Modify the path accordingly)
df = pd.read_csv("/content/drive/MyDrive/Sem-6/ML/Lab1/sales_data.csv")

# Display first few rows to check column names
print(df.head())

# Select independent variables (X) and dependent variable (Y)
# Automatically selecting all columns except target column (assumes last column is target)
X = df.iloc[:, :-1].values  # Independent variables (all columns except last)
Y = df.iloc[:, -1].values   # Dependent variable (last column)

# Train the Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict the output
Y_pred = model.predict(X)

# Print Model Parameters
print(f"Intercept: {model.intercept_:.2f}")
print(f"Coefficients: {model.coef_}")

# Calculate errors
mae = mean_absolute_error(Y, Y_pred)
mse = mean_squared_error(Y, Y_pred)
rmse = np.sqrt(mse)

print(f"Mean Absolute Error: {mae:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")

# Plot actual vs predicted values (Only applicable for 1D or summary visualization)
plt.figure(figsize=(8, 5))
plt.scatter(Y, Y_pred, color='blue', label="Actual vs Predicted")
plt.plot([min(Y), max(Y)], [min(Y_pred), max(Y_pred)], color='red', linewidth=2, label="Regression Line")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Linear Regression: Actual vs Predicted")
plt.legend()
plt.grid(True)
plt.show()
```
